## create own aggregation

```python
def max_to_min(arr):
    return arr.max()-arr.min()
```

```python
wino = dframe_wine.groupby('quality')
```

```python
wino.describe()
```

## To apply aggregate function use .agg

### 먼저 그룹한 후에 함수를 적용!

### wino 는 'quality'를 기준으로 이미 그룹된 변수

```python
wino.agg(max_to_min)
```

```python
wino.agg('mean')
```
