```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
dframe_wine = pd.read_csv('winequality-red.csv',sep=';')
dframe_wine.head()
```

```python
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

## change dataframe 'inplace=True'

```python
dframe_wine.sort('alcohol',ascending = False,inplace=True)
```

### now we have the dataframe sorted

```python
dframe_wine = dframe_wine.groupby('quality').apply(ranker)
```

```python
dframe_wine.head(10)
```

```python
dframe_wine.tail(10)
```

## NOW finding highest alcohol content for each ranking..!

```python
num_of_qual = dframe_wine['quality'].value_counts()
num_of_qual
```
